```python
# reading in CSV with data on shortage of care

df = pd.read_csv('Overall Supply and Estimated Demand for Child Care by Age and By Town.csv')
df = df.replace(',','', regex=True)

df.head()
```

```python
# delete first row in df, which is total of all the towns in df

df_noneRow1 = df.drop(0)
df_noneRow1
```

```python
# converting the Shortage_of_care_InfantsToddlers column to numeric so that sorting and analysis can be done
# this was done as a test, not on the original df

df_noneRow1['Shortage_of_care_InfantsToddlers'] = pd.to_numeric(df_noneRow1['Shortage_of_care_InfantsToddlers'])
```

```python
# Sorted descending list for Shortage_of_care_InfantsToddlers

need_met_sorted_no1 = df_noneRow1['Shortage_of_care_InfantsToddlers'].sort_values(ascending=False)
need_met_sorted_no1
```

```python
# converting the Shortage_of_care_Preschoolers column to numeric so that sorting and analysis can be done
# same process as before, except done to the main df

df['Shortage_of_Care_Preschoolers'] = pd.to_numeric(df['Shortage_of_Care_Preschoolers'])
```

```python
# sorted descending list to look for any irregularities

need_met_sorted2 = df['Shortage_of_Care_Preschoolers'].sort_values(ascending=False)
need_met_sorted2
```

```python
# now the Shortage_of_care_InfantsToddlers column is converted to numeric within the df
# this will be analyzed

df['Shortage_of_care_InfantsToddlers'] = pd.to_numeric(df['Shortage_of_care_InfantsToddlers'])
```

```python
# another instance of converting Shortage_of_Care_Preschoolers column to numeric
# in the case that you want to skip creating a sorted list to check if they are numeric

df['Shortage_of_Care_Preschoolers'] = pd.to_numeric(df['Shortage_of_Care_Preschoolers'])
```

```python
# sorted subset of top 30 towns for Shortage of care for Infants and toddlers

df2 = df.sort_values(['Shortage_of_care_InfantsToddlers']).head(30)
df2 = df2.drop(0)

df2
```

```python
# produce a plot of shortage of care for infants and toddlers from the sorted subset df2
# ascending sort of shortage of care for infants and toddlers

fig = px.bar(df2, x="City", y="Shortage_of_care_InfantsToddlers", title='Top 30 towns for shortage of infant and toddler care') 
fig.update_xaxes(tickangle=45)
fig.show()
```

```python
# produce a plot of shortage of care for preschoolers from the same sorted subset df2
# based on the ascending sort of shortage of care for infants and toddlers

fig = px.bar(df2, x="City", y="Shortage_of_Care_Preschoolers", title='Shortage of preschooler care for the same 30 towns') 
fig.update_xaxes(tickangle=45)
fig.show()
```

```python
# sorted subset of top 30 towns for Shortage of care for preschoolers

df3 = df.sort_values(['Shortage_of_Care_Preschoolers']).head(30)

df3
```

```python
# produce a plot of shortage of care for preschoolers from the sorted subset df3
# ascending sort of shortage of care for preschoolers

fig = px.bar(df3, x="City", y="Shortage_of_Care_Preschoolers", title='Top 30 towns for shortage of preschooler care') 
fig.update_xaxes(tickangle=45)
fig.show()
```

```python
# produce a plot of shortage of care for infants and toddlers from the same sorted subset df3
# based on the ascending sort of shortage of care for preschoolers
# in this case, it looks like shortage of care for preschoolers guarantees shortage of care for infants and toddlers

fig = px.bar(df3, x="City", y="Shortage_of_care_InfantsToddlers", title='Shortage of infant and toddler care for the same 30 towns') 
fig.update_xaxes(tickangle=45)
fig.show()
```

```python
# Additional analysis that can be connected to the previous town based analysis
# PUMA region dataset

# read in PUMA region csv
# performing some regex in order to make the columns

puma = pd.read_csv('Average, Annual, Advertised Price by Age, Setting Type, and PUMA Region.csv')
puma = puma.replace(',','', regex=True)
puma = puma.replace('\$','', regex=True)

puma.head()
```

```python
# ascending sort of this dataset based on Overall_Average_Rate_InfantsToddlers column

# statewide average for overall cost rate for infant and toddler care care: $13,780

puma_2 = puma.sort_values(['Overall_Average_Rate_InfantsToddlers'])

puma_2
```
